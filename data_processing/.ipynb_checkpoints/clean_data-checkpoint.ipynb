Use regex to get strains we're interested in

```python
def regex_selector(name):
#     print(name)
    if type(name) != str: return False
    return bool(re.search('H1N1|H3N2|H5N1|H7N3', name))
```

```python
fludf = fludf[fludf['Pathogen'].apply(regex_selector)]
```
