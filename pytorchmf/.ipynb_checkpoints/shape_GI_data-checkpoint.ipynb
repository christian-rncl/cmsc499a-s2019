```python
import pandas as pd
```

```python
df = pd.read_csv('zscores - z-scores.csv')
```

```python
df.head()
```

```python
df = df.drop(['mut'], axis=1)
```

```python
df.head()
```

```python
i, row = next(df.iterrows())
print(row['gene'])
```

```python
effectors = df.columns[1:].values

newdf = {'gene':[], 'effector':[], 'zscore':[]}
j = 0
for _, row in df.iterrows():
    gene = row['gene']
    
    for e in effectors:
        newdf['gene'].append(gene)
        newdf['effector'].append(e)
        newdf['zscore'].append(row[e])
```

```python
newdf = pd.DataFrame(newdf)
```

```python
newdf.head()
```

```python
newdf.to_csv('gi_dset.csv', index=False)
```
